```python
import cv2
```

```python
cascade_face= cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap= cv2.VideoCapture(0)

while True:
    rect, img=cap.read()  #rect is for rectangle when it recognizes face.
    print(rect)  
    g=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)   #g is for making image in grayscale color.
    
    #for multiscale faces, cascade face is used on the grayscale image by giving scale size    
    f=cascade_face.detectMultiScale(g, 1.3, 6)# 4 same in cv2.rectangle of for loop.
    
    for (x, y, w, h) in f:   # x-axis y-axis width & height in f
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 255, 0), 6) #BLUE,GREEN,RED COLOR IN 3RD bracket & 4 is width. 
        
   #x & w lie ->horizontal, y & h lie -> vertical,'B G R' color  TO show any color , put 255, and keep other two as 0
        
    cv2.imshow('img',img)
    
    k=cv2.waitKey(30) & 0xff 
    #wait for (15 sec with AND of 0xff(hexadecimal constant))

    if k==27:
        break

cap.release()
cap.destroyAllWindows()
    
```
